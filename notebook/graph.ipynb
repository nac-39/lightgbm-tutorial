```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import rcParams
from matplotlib.font_manager import FontProperties

# 
font_path = '/System/Library/Fonts/ W3.ttc'
font_prop = FontProperties(fname=font_path)
rcParams['font.family'] = font_prop.get_name()


## 0.5ShanonGini
def entropy(p):
    return -0.5 * (p * np.log2(p) + (1 - p) * np.log2(1 - p))


def gini(p):
    return 2 * p * (1 - p)


def misclassification(p):
    return 1 - np.maximum(p, 1 - p)


p = np.linspace(0, 1, 100)

plt.figure(figsize=(10, 6))
plt.plot(p, entropy(p), label='0.5Shanon')
plt.plot(p, gini(p), label='Gini')
plt.plot(p, misclassification(p), label='')
plt.xlabel(' p')
plt.ylabel('')
plt.legend()
plt.show()

```
