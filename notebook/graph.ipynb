```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import rcParams
from matplotlib.font_manager import FontProperties

# 日本語フォントのパスを設定
font_path = '/System/Library/Fonts/ヒラギノ角ゴシック W3.ttc'
font_prop = FontProperties(fname=font_path)
rcParams['font.family'] = font_prop.get_name()


## 0.5×Shanonエントロピー、Gini不純度、誤分類率のグラフを書く
def entropy(p):
    return -0.5 * (p * np.log2(p) + (1 - p) * np.log2(1 - p))


def gini(p):
    return 2 * p * (1 - p)


def misclassification(p):
    return 1 - np.maximum(p, 1 - p)


p = np.linspace(0, 1, 100)

plt.figure(figsize=(10, 6))
plt.plot(p, entropy(p), label='0.5×Shanonエントロピー')
plt.plot(p, gini(p), label='Gini不純度')
plt.plot(p, misclassification(p), label='誤分類率')
plt.xlabel('確率 p')
plt.ylabel('値')
plt.legend()
plt.show()

```
